Rename Index Labels or Columns in a DataFrame

```python
data =data.rename(columns={'Year': 'Year of Release', 'Box Office':'Revenue'})
data.head()
```

```python
#You can set up a dictionary and later change the values in the data frame
swaps = {
    "Dr. No": "Dr No",
    "GoldenEye": "Golden Eye",
    "The World Is Not Enough": "Best Bond Movie Ever"
}
data=data.rename(index=swaps)
data.head()
```

The apply Method with DataFrames

```python
data = pd.read_csv("jamesbond.csv", index_col="Film").sort_index()
```

```python
def rank_movie(row):
    year = row.loc['Year']
    actor = row.loc['Actor']
    budget = row.loc['Budget']

    if year >= 1980 and year <1990:
        return 'Great 80s movie'
    if actor == "Pierce Brosnan":
        return "The best Bond ever!"
    if budget > 100:
        return "Expensive movie, fun"

    return "No comment"

data.apply(rank_movie, axis='columns')
```
